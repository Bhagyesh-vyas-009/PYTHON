```python
import requests
import json

# Replace 'YOUR_API_KEY' with your actual TMDB API key
API_KEY = 'cf20abe739a83491bf539c8477df6944'
BASE_URL = 'https://api.themoviedb.org/3/discover/movie'

def fetch_movies(params, page=1):
    params['page'] = page
    response = requests.get(BASE_URL, params=params)
    response.raise_for_status()  # Raise an error for bad status codes
    return response.json()

def main():
    # Define the parameters
    params = {
        'api_key': API_KEY,
        'language': 'en-US',                # Language of the results
        'sort_by': 'popularity.desc',       # Sorting method
        'include_adult': 'false',           # Include adult movies
        'include_video': 'false',           # Include movies with video
        'with_genres': '',                  # Filter by genre IDs (comma separated)
        'with_cast': '',                    # Filter by cast IDs (comma separated)
        'with_crew': '',                    # Filter by crew IDs (comma separated)
        'with_companies': '',               # Filter by production company IDs (comma separated)
        'with_keywords': '',                # Filter by keyword IDs (comma separated)
        'with_watch_providers': '',         # Filter by watch provider IDs (comma separated)
        'watch_region': '',                 # Specify a watch region
        'release_date.gte': '',             # Release date greater than or equal to (YYYY-MM-DD)
        'release_date.lte': '',             # Release date less than or equal to (YYYY-MM-DD)
        'vote_average.gte': '',             # Minimum vote average
        'vote_average.lte': '',             # Maximum vote average
        'vote_count.gte': '',               # Minimum vote count
        'vote_count.lte': '',               # Maximum vote count
        'with_runtime.gte': '',             # Minimum runtime (minutes)
        'with_runtime.lte': '',             # Maximum runtime (minutes)
        'primary_release_year': '',         # Primary release year
        'primary_release_date.gte': '',     # Primary release date greater than or equal to (YYYY-MM-DD)
        'primary_release_date.lte': '',     # Primary release date less than or equal to (YYYY-MM-DD)
        'year': '',                         # Filter by year
        'region': '',                       # Specify a region
        'certification_country': '',        # Certification country
        'certification': '',                # Certification
        'certification.lte': '',            # Certification less than or equal to
        'with_release_type': '',            # Filter by release type
        'with_original_language': ''        # Filter by original language
    }

    # Optionally, update the parameters with specific values
    # params.update({
    #     'with_genres': '28,12',  # Example: Action and Adventure genres
    #     'primary_release_year': '2022',
    #     'vote_average.gte': '7.5'
    # })

    all_movies = []
    page = 1
    total_pages = 1

    while page <= total_pages:
        data = fetch_movies(params, page)
        all_movies.extend(data['results'])
        total_pages = data['total_pages']
        page += 1

    # Create a JSON document of all movies
    with open('all_movies.json', 'w') as f:
        json.dump(all_movies, f, indent=4)

    print(f'Total movies fetched: {len(all_movies)}')

if __name__ == '__main__':
    main()
```
