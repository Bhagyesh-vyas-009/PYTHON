### 01) WAP to demonstrate the use of Pie chart.

```python
import matplotlib.pyplot as plt
%matplotlib inline
values=[40,35,10,15]
l=['A','B','C','D']
c=['b','r','g','y']
plt.pie(values,labels=l,colors=c)
plt.show()
```

### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

```python
import random
x=[10,20,30,40,50]
y=[random.randint(1,50) for i in range(5)]
plt.plot(x,y,c="m",ls=':',marker='>')
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
%matplotlib inline
x=[1,2,3,4,5]
y=[2.5,5.6,7,8.7,9.55]
l=['1st','2nd','3rd','4th','5th']
c=['b','r','g','y','c']
w = [0.5,0.6,0.3,0.8,0.9]
plt.title('Sem wise spi')
plt.bar(x,y,color=c,label=l,width=w)
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
%matplotlib inline
x=["python","c","c++","java"]
y=[85,70,60,80]
c=["r","g","y","m"]
plt.bar(x,y,color=c,align='edge',edgecolor='b',alpha=0.5,)
plt.title("Popularity of langauges")
plt.xlabel("langauges")
plt.ylabel("popularity")
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
addlabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=["r","g","y","m","b"]
s=[150,70,190,50,120]
plt.scatter(x,y,color=c,marker='*',s=s)
plt.title("CPI && SPI")
plt.xlabel("CPI")
plt.ylabel("SPI")
plt.show()
```
